## 4.1. Import packages and load data

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
directory = '../../Data/tidy_data/'
```

```python
# During hypo 5
#during_glyc = pd.read_csv(directory+'glyc_during.csv')
during_hypo_5 = pd.read_csv(directory+'prediction_df_during_hypo_5.csv')
# During hypo 3
during_hypo_3 = pd.read_csv(directory+'prediction_df_during_hypo_3.csv')
# During hyper
during_hyper = pd.read_csv(directory+'prediction_df_during_hyper.csv')

# 1hr after
#one_hr_glyc = pd.read_csv(directory+'glyc_1hr_after.csv')
one_hr_glyc_ts = pd.read_csv(directory+'prediction_df_1hr_after.csv')
# 4hrs after
#four_hr_glyc = pd.read_csv(directory+'glyc_4hr_after.csv')
four_hr_glyc_ts = pd.read_csv(directory+'prediction_df_4hr_after.csv')
```

## 4.2. Drop and tidy columns

```python
during_hypo_5[['ID', 'bout_id', 'start_datetime','finish_datetime', 'study']].to_csv('id_key.csv', index=False)
```

```python
# Columns to drop
all_drop = ['study', 'start_datetime', 'finish_datetime', 'type_of_exercise',
             'starting_glucose', 'finishing_glucose']
lab_drop = ['insulin_administration', 'urine_cpep', 'urine_creat', 'chol',
             'h_index', 'hdl', 'ldl', 'nhdl', 'trig']
during_drop = ['end_glc', 'end_roc']
```

```python
# Dropping columns for during datasets
during_frames = [during_hypo_5, during_hypo_3, during_hyper]
for i in during_frames:
    i.drop(columns=all_drop+lab_drop+during_drop, inplace=True)
```

```python
# Dropping columns for after datasets
after_frames = [one_hr_glyc_ts, four_hr_glyc_ts]
for i in after_frames:
    i.drop(columns=all_drop+lab_drop, inplace=True)
```

```python
during_hypo_5.merge(during_hypo_3, on='bout_id')[['bout_id','start_glc_x','duration_x', 'y_hypo','y_hypo_3']].sort_values('start_glc_x')
```

```python
#during_hypo_5['y_hypo'] = during_hypo_5['y_hypo'] > 0
during_hypo_5['stratify'] = during_hypo_5.y_hypo.astype(str) + '_'+ during_hypo_5.ID.astype(str)

#during_hypo_3['y_hypo_3'] = during_hypo_3['y_hypo_3'] > 0
during_hypo_3['stratify'] = during_hypo_3.y_hypo_3.astype(str) + '_'+ during_hypo_3.ID.astype(str)
#during_hyper['y_hyper'] = during_hyper['y_hyper'] > 0
during_hyper['stratify'] = during_hyper.y_hyper.astype(str) + '_'+ during_hyper.ID.astype(str)

#one_hr_glyc_ts['y_hypo'] = one_hr_glyc_ts['y_hypo'] > 0
one_hr_glyc_ts['stratify'] = one_hr_glyc_ts.y_hypo.astype(str) + '_'+ one_hr_glyc_ts.ID.astype(str)

#four_hr_glyc_ts['y_hypo'] = four_hr_glyc_ts['y_hypo'] > 0
four_hr_glyc_ts['stratify'] = four_hr_glyc_ts.y_hypo.astype(str) + '_'+ four_hr_glyc_ts.ID.astype(str)
```
